```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('netflix_data.csv')
print(df.head())
```

```python
df=df.dropna(subset=['release_year'])   #clean the data
```

```python
df['release_year']=df['release_year'].astype(int)
```

```python
plt.figure(figsize=(10,5))
plt.hist(df['release_year'],bins=50,color='hotpink',edgecolor='darkblue',alpha=0.8)
plt.title('Netflix released Titles by Year')
plt.xlabel('Release Year')
plt.ylabel('No of Titles')
plt.grid(color='grey',linestyle='--',linewidth=0.5)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('netflix_data.csv')
df=df.dropna(subset=['release_year'])   #clean the data
df['release_year']=df['release_year'].astype(int)
plt.figure(figsize=(8,5))
plt.hist(df['release_year'],bins=50,color='green',edgecolor='yellow',alpha=0.8)
plt.title('Netflix released Titles by Year')
plt.xlabel('Release Year')
plt.ylabel('No of Titles')
plt.grid(color='grey',linestyle='--',linewidth=0.5)
plt.tight_layout()
plt.savefig('netflix_bar.pdf',dpi=300,bbox_inches='tight')
plt.show()
```

```python
df=df.dropna(subset=['release_year','type'])
df['release_year']=df['release_year'].astype(int)
```

```python
movies=df[df['type']=='Movie']
shows=df[df['type']=='TV Shows']
```

```python
plt.figure(figsize=(10,5))
plt.hist([movies['release_year'],
         shows['release_year']],
         bins=40,
         color=['green','red'],
         edgecolor='yellow',
         alpha=0.8,
         label=['Movies','TV Shows']
        )
plt.title('Netflix released Titles by Year and Type')
plt.xlabel('Release Year')
plt.ylabel('No of Titles Released')
plt.grid(axis='y',color='grey',linestyle='--',linewidth=0.5)
plt.tight_layout()
plt.savefig('netflix_showsmovies.jpg',dpi=300,bbox_inches='tight')
plt.legend()
plt.show()
```

```python
df=df.dropna(subset=['type'])   #clean the data
```

```python
type_counts=df['type'].value_counts()
print(type_counts)
```

```python
plt.figure(figsize=(7,7))
plt.pie(type_counts,
        labels=type_counts.index,
        colors=plt.cm.inferno([0.1,0.5]),
        textprops={'color':'white','fontsize':10,'fontweight':'bold'},
        autopct='%1.2f%%',
        startangle=140,
        shadow=True,
        explode=(0.05,0.03))
plt.title('Netflix Movies/TV Shows Distribution')
plt.savefig('netflix_pieplot2.pdf',dpi=300,bbox_inches='tight')
plt.legend(title='Movies & Shows')
plt.show()
```
