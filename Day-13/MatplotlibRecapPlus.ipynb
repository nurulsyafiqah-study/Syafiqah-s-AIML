```python
import matplotlib.pyplot as plt
```

```python
months=['Jan','Feb','Mar','April']
sales=[2000,3000,2500,3600]
```

### plt.plot(x,y,color='color name', linestyle='line_style', linewidth=value, marker='marker symbol', label='label name'

```python
plt.plot(months,sales,color='hotpink', linestyle='--', linewidth=2, marker='o',label='month sales')
plt.xlabel('Months')
plt.ylabel('Sales per Month')
plt.title('Monthly Sales Report')
plt.xlim(0,4)
plt.ylim(2000,5000)
plt.grid(True)
plt.legend(loc=('lower right'))
plt.show()
```

```python
import numpy as np
```

```python
x=np.linspace(0,2,100)
plt.plot(x,x,label='Linear')
plt.plot(x,x**2,label='Quadratic')
plt.plot(x,x**3,label='Cubic')
plt.xlabel('X Lines')
plt.ylabel('Y Lines')
plt.title('Simple Line Plot')
plt.legend()
plt.show()
```

# Bar Plot Example
## plt.bar(x,y,color=",edgecolor='')

```python
categories=['Electronics','Mobiles','Books','Groceries','Home Decoration']
sales=[50000,60000,30000,90000,40000]
bar_colors=['pink','black','purple','cyan','orange']
plt.figure(figsize=(8,5))
plt.bar(categories,sales,color=bar_colors,edgecolor='yellow')
plt.xlabel('Categories')
plt.ylabel('Sales Amount (RM)')
plt.title('Sales Report-Categories')
plt.grid(axis='y',linestyle='--',color='gray',alpha=0.5)
plt.tight_layout()
plt.show()
```

```python
categories=['Electronics','Mobiles','Books','Groceries','Home Decoration']
sales=[50000,60000,30000,90000,40000]
bar_colors=['pink','black','purple','cyan','orange']
plt.figure(figsize=(8,5))
plt.bar(categories,sales,color=bar_colors,edgecolor='yellow',label='Sales Info')
plt.xlabel('Categories')
plt.ylabel('Sales Amount (RM)')
plt.title('Sales Report-Categories')
plt.grid(axis='y',linestyle='--',color='gray',alpha=0.5)
plt.xticks(rotation=30,fontsize=12,color='magenta')
plt.yticks([20000,40000,60000,80000,100000],['20K','40K','60K','80K','1M'],fontsize=12,color='green')
plt.legend(title='Legend',fontsize=6)
plt.tight_layout()
plt.show()
```

# Scatter Plot
###### plt.scatter(x,y,s=value,c='colorName',marker='markerName',alpha=transp.,edgecolor='colorName',label='label of Legend')

```python
height_cm=[150,155,160,165,170,175,180,185]
weight_kg=[50,53,60,59,68,62,72,80]
plt.figure(figsize=(8,5))
plt.scatter(height_cm,weight_kg,s=100,c='pink',marker='*',alpha=0.9,edgecolor='black',label='weight-height comparison')
plt.title('Scatter Plote Height vs Weight')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.grid(color='lightblue',alpha=0.5,linestyle='--',linewidth=0.5)
plt.xticks(rotation=30,fontsize=12,color='black')
plt.yticks([50,55,60,65,70,75,80,85],['50','55','60','65','70','75','80','85'],fontsize=12,color='black')
plt.legend()
plt.show()
```
