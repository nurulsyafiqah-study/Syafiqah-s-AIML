# First Example to create a line chart from our data
    ## students.xlsx for fething the data
    ## Generate Line Chart based on Student Name and Student Marks

```python
import pandas as pd
```

```python
df=pd.read_excel('C:/Users/Syafiqah/OneDrive/Desktop/AIML/Day-10/students.xlsx')
df
```

```python
names=df['Name']
marks=df['Marks']
```

```python
plt.plot(names,marks,marker='o',color='blue')
plt.title('Student Marks Line Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(4,2))
plt.plot(names,marks,marker='s',color='pink')
plt.title('Student Marks Line Chart')
plt.xlabel('Student Name')
plt.ylabel('Student Marks')
plt.grid(True)
plt.show()
```

### #with dotted line

```python
plt.plot(names,marks,marker='p',color='blue',linestyle='--')
plt.show()
```
