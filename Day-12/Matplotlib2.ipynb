```python
players=['MSD','Rohit','Virat','Gill','Shami']
match1_score=[45,67,56,76,89]
match2_score=[55,78,40,80,97]
match3_score=[90,40,20,30,60]
plt.figure(figsize=(8,5))
plt.scatter(players,match1_score,color='red',label='Match One',s=100)
plt.scatter(players,match2_score,color='gray',label='Match Two',s=100,marker='x')
plt.scatter(players,match3_score,color='blue',label='Match Three',s=100,marker='v')
plt.title('Comparison of Players in Two Matches')
plt.xlabel('Players')
plt.ylabel('Score')
plt.legend(loc='best') #or can use: lower left,centre,upper right, whatever
plt.grid(True,linestyle='--',alpha=0.5)
plt.show()
```

# Create a vertical bar graph where each bar represents the value of its respective liked by users
    ## every bar must have different color


#### #Vertical bar graph

```python
python_lib=['Numpy','Pandas','Matplotlib','Scikit-learn']
liked_by=[5000,7000,9000,4000]
plt.figure(figsize=(6,2))
bar_color=['turquoise','pink','cyan','yellow']
plt.bar(python_lib,liked_by,color=bar_color,width=1,edgecolor='red')
plt.title('Popularities of Python Libraries Among Users')
plt.xlabel('Python Library')
plt.ylabel('Likes Count')
plt.show()
```

#### #Horizontal bar graph

```python
plt.figure(figsize=(6,2))
plt.barh(python_lib,liked_by,color=bar_color,edgecolor='red')
plt.title('Popularities of Python Libraries Among Users')
plt.xlabel('Likes Count')
plt.ylabel('Python Library')
plt.show()
```
