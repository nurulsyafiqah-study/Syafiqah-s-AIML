```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.random.rand(50)
y=np.random.rand(50)
temp=np.random.randint(20,100,size=50)
plt.scatter(x,y,c=temp,cmap='plasma',s=120,edgecolor='black')
plt.colorbar(label='Temperature')
plt.title('Scatter Plot Colored by Temperature')
plt.xlabel('X-Random')
plt.ylabel('Y-Random')
plt.show()
```

```python
height_cm=[150,155,160,165,170,180,185]
weight_kg=[50,53,56,59,62,66,68]
age_year=[20,22,25,28,30,34,38]
```

```python
plt.figure(figsize=(8,5))
```

```python
scatter=plt.scatter(height_cm,weight_kg,c=age_year,cmap='rainbow',s=120,alpha=0.9)
plt.colorbar(scatter,label='Age')
plt.title('Scatter Plot: Height vs Weight (colored by age)')
plt.xlabel('Height(cm)')
plt.ylabel('Weight(kg)')
plt.grid(color='blue',alpha=0.5,linestyle='--',linewidth=0.5)
plt.show()
```

# Pie Plot Example
#### plt.pie(x,labels'',colors='',autopct='',startagnle='',shadow=,explode=,radius)

```python
dept=['HR','Web-Development','App-Development','Agile-Scrum','Testing']
expenses=[12000,25000,18000,15000,20000]
```

```python
plt.figure(figsize=(7,7))
```

```python
plt.pie(expenses,labels=dept,autopct='%1.2f%%',startangle=140,explode=(0,0.05,0,0,0))
plt.title('Department wise Expenses Report')
plt.show()
```

#### COOL color pie chart

```python
plt.pie(expenses,labels=dept,colors=plt.cm.cool([0.1,0.3,0.5,0.7,0.9]),autopct='%1.2f%%',startangle=140,explode=(0,0.05,0,0,0))
plt.title('Department wise Expenses Report')
plt.show()
```

#### RAINBOW color pie chart

```python
plt.pie(expenses,labels=dept,colors=plt.cm.rainbow([0.1,0.3,0.5,0.7,0.9]),autopct='%1.2f%%',startangle=140,explode=(0,0.05,0,0,0))
plt.title('Department wise Expenses Report')
plt.show()
```

#### HOT color pie chart

```python
plt.pie(expenses,labels=dept,colors=plt.cm.hot([0.1,0.3,0.5,0.7,0.9]),autopct='%1.2f%%',startangle=140,explode=(0,0.05,0,0,0))
plt.title('Department wise Expenses Report')
plt.show()
```

#### PLASMA color pie chart

```python
plt.pie(expenses,labels=dept,colors=plt.cm.plasma([0.1,0.3,0.5,0.7,0.9]),autopct='%1.2f%%',startangle=140,explode=(0,0.05,0,0,0))
plt.title('Department wise Expenses Report')
plt.show()
```

#### INFERNO color pie chart

```python
plt.pie(expenses,labels=dept,colors=plt.cm.inferno([0.1,0.3,0.5,0.7,0.9]),autopct='%1.2f%%',startangle=140,explode=(0,0.05,0,0,0))
plt.title('Department wise Expenses Report')
plt.show()
```

# Create a pie plot using following data
    # sales_data={'Mobile':45000,'Clothing':30000,'Groceries':35000,'Home Decor':20000}
    # show percentage with two decimal
    # use colormap to give the different color to your pie

```python
sales_data={'Mobile':45000,'Clothing':30000,'Groceries':35000,'Home Decor':20000}
item=list(sales_data.keys())
sales=list(sales_data.values())
plt.figure(figsize=(7,7))
plt.pie(sales,labels=item,colors=plt.cm.inferno([0.1,0.3,0.5,0.7]),textprops={'color':'white','fontsize':10,'fontweight':'bold'},autopct='%1.2f%%',startangle=140,explode=(0.05,0,0,0))
plt.title('Sales Report')
#plt.savefig('filenamewithextension',dpi=value,bbox_inches='')
plt.legend()
plt.show()
```

```python
sales_data={'Mobile':45000,'Clothing':30000,'Groceries':35000,'Home Decor':20000}
item=list(sales_data.keys())
sales=list(sales_data.values())
plt.figure(figsize=(7,7))
plt.pie(sales,labels=item,colors=plt.cm.inferno([0.1,0.3,0.5,0.7]),textprops={'color':'white','fontsize':10,'fontweight':'bold'},autopct='%1.2f%%',startangle=140,explode=(0.05,0,0,0))
plt.title('Sales Report')
#plt.savefig('filenamewithextension',dpi=value,bbox_inches='')
plt.savefig('Categories wise Sales Report',dpi=300,bbox_inches='tight')
plt.legend()
plt.show()
```

# Histogram Example

```python
marks=[45,60,80,55,62,90,89,50,56,60,80,55,62,90,89,58,90,56,67,88]
#print(len(marks)) - to count the data that were put in marks
#plt.style.use('ggplot')
plt.figure(figsize=(8,5))
plt.hist(marks,bins=5,color='blue',edgecolor='darkblue',alpha=0.8)
plt.title('Histogram for Student Marks')
plt.xlabel('Marks Range')
plt.ylabel('No of Students')
plt.grid(color='red',linestyle='--',linewidth=0.5)
plt.show()
```
